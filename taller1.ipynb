```r
# CONTIGENCY TABLE AND FRECUENCIES
# Marriages and having children 

install.packages("wooldridge")
install.packages("RColorBrewer")
```

```r
# Call the library and data
library(RColorBrewer)
library(wooldridge)
data(affairs, package = "wooldridge")
```

```r
# Qualification of persons having children
haskid <- c("no", "yes")
kids <- factor(affairs$kids, labels = haskid)

# Call data rating happiness of Marriage 
mlab <- c("very unhappy", "unhappy", "averge", "happy", "very happy")
marriage <- factor(affairs$ratemarr, labels = mlab)

countslab <- table(marriage, kids)
percentage <- prop.table(countslab, margin = 1)
percentage2 <- prop.table(countslab, margin = 2)

#  Pie Graph
percent <- round(c(prop.table(table(marriage))*100), digits = 2)


color <- brewer.pal(length(mlab), "Accent")
par(mai = c(0,0,0,0))
pie(table(marriage), 
  col = color, 
  border = "white",
  labels = paste(percent, "%"), 
  radius = 0.6,)
  title("Rating of marriage happiness",
    line = -4)
  legend(
    -0.95,-0.85,
    legend = c("very unhappy", "unhappy", "averge", "happy", "very happy"),
    horiz = TRUE,
    col = color,
    fill = color,
    cex = 0.8)
```

```r
# Graph distribution of happiness
par(mai = c(0.7,1.4,0.9,0.5))
barplot(table(marriage),
  horiz = TRUE, las = 1,
  col = color)
  title("Distribution of Happines")
  legend(
    170, 2.3,
    legend = table(marriage),
    col = color,
    fill = color,
    cex = 1.2)
```

```r
# Distribution of marriages with children
percent2 <- prop.table(table(kids))*100
p = round(percent2, digits = 2)
color2 <- brewer.pal(length(haskid), "Set1")
par(mai = c(0,0,0,0))
pie(table(kids),
  col = color2,
  border = "white",
  labels = paste(p, "%"),
  radius = 0.6,)
  title("Distribution of marriages with children",
    line = -4)
  legend(
    0.6, -0.7,
    legend = c("no","yes"),
    col = color,
    fill = color2,
    cex = 1.3)
```

```r
barplot(table(kids),
  col = color2,
  main = "Distribution of marriages with children")
  legend("topleft",
    legend = table(kids),
    col = color2,
    fill = color2)
```

```r
color3 <- brewer.pal(length(mlab), "Paired")
par(mai = c(1,1.5,1,1))
barplot(table(kids, marriage),
  col = color3,
  xlim = c(0, 250),
  xlab = "Quantity",
  horiz = TRUE, las = 1)
  title("Happiness by kids")
  legend(
    170, 2.3,
    legend = c("no", "yes"),
    col = color3,
    fill = color3,
    cex = 1.3)
```

```r
barplot(prop.table(table(kids, marriage), margin = 2),
  col = color3,
  main = "Happiness by kids",
  ylab = "Percentage")
  legend(
    3.6,0.94,
    horiz = TRUE,
    legend = c("no", "yes"),
    col = color3,
    fill = color3,
    cex = 1.3)
```
